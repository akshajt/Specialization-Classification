### Visualising Data

```python
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(data= fetus, x="fetal_health")
```

```python
plt.figure(figsize = (20,20))
sns.heatmap(fetus.corr(), annot=True, cmap = 'YlGnBu')
plt.show()
```

```python
fetus.hist(figsize = (30,30))
plt.show()
```

```python
X=fetus.drop(['fetal_health'], axis=1)
y=fetus['fetal_health']
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix,plot_confusion_matrix, plot_roc_curve
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 42)
```

```python
from sklearn.feature_selection import SelectKBest
bestFeatures = SelectKBest(k='all')
fit = bestFeatures.fit(X_train,y_train)

scores = pd.DataFrame(fit.scores_)
columns = pd.DataFrame(X.columns)
featureScores = pd.concat([columns, scores], axis=1)
featureScores.columns = ['Feature','Score']

plot = sns.barplot(data=featureScores, x='Score', y='Feature', palette='plasma')
plt.box(False)
plot
```

#### Considering only the columns with a K score over 50

```python
selected = featureScores[featureScores['Score']>=50]
selected = list(selected['Feature'])
selected.append('fetal_health')
fetus = fetus[selected]
fetus.head()
```

### Using a Standard Scaler

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## KNN

```python
knn_clf = KNeighborsClassifier().fit(X_train,y_train)
y_pred_knn = knn_clf.predict(X_test)
print(classification_report(y_test, y_pred_knn))
```

## Logistic Regression

```python
lr_clf = LogisticRegression(random_state=42).fit(X_train,y_train)
y_pred_lr = lr_clf.predict(X_test)
print(classification_report(y_test, y_pred_lr))
```

## Kernelized Support Vector Machine

```python
sv_clf = SVC(random_state=42).fit(X_train, y_train)
y_pred_sv = sv_clf.predict(X_test)
print(classification_report(y_test, y_pred_sv))
```

## Decision Tree Classifier

```python
dt_clf = DecisionTreeClassifier(random_state=42).fit(X_train, y_train)
y_pred_dt = dt_clf.predict(X_test)
print(classification_report(y_test, y_pred_dt))
```

```python
print("Accuracy: ", accuracy_score(y_test, y_pred_dt))
print("Precision: ", precision_score(y_test, y_pred_dt, average="weighted"))
print("Recall: ", recall_score(y_test, y_pred_dt, average="weighted"))
print("F1 Score: ", f1_score(y_test, y_pred_dt, average="weighted"))
```

## Random Forest Classifier

```python
rf_clf = RandomForestClassifier(random_state=42).fit(X_train, y_train)
y_pred_rf = rf_clf.predict(X_test)
print(classification_report(y_test, y_pred_rf))
```

```python
print("Accuracy: ", accuracy_score(y_test, y_pred_rf))
print("Precision: ", precision_score(y_test, y_pred_rf, average="weighted"))
print("Recall: ", recall_score(y_test, y_pred_rf, average="weighted"))
print("F1 Score: ", f1_score(y_test, y_pred_rf, average="weighted"))
```

## Naive Bayes Classifier

```python
nb_clf = GaussianNB().fit(X_train, y_train)
y_pred_nb = nb_clf.predict(X_test)
print(classification_report(y_test, y_pred_nb))
```

#### Decision Tree Classifier, Random Forest Classifier, and SVC produced the best results. 
#### We will therefore try to optimize these 3 algorithms by employing a grid search to tune the hyperparameters

## Tuned Random Forest

```python
grid_params_rf={
    "criterion":["gini", "entropy"],
    "n_estimators":[50, 100, 150, 200, 250],
    "max_depth":[6, 8, 10, 12, 14],
    "min_samples_split":[2,4,6]
}
```

```python
gs_rf = GridSearchCV(RandomForestClassifier(), grid_params_rf, verbose=1, n_jobs=-1)
gs_rf.fit(X_train, y_train)
```

```python
gs_rf.best_params_
```

```python
randomForest = RandomForestClassifier(**gs_rf.best_params_).fit(X_train, y_train)
y_pred = randomForest.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred, average="weighted"))
print("Recall: ", recall_score(y_test, y_pred, average="weighted"))
print("F1 Score: ", f1_score(y_test, y_pred, average="weighted"))
```

```python
plot_confusion_matrix(randomForest, X_test, y_test, display_labels = ['Normal', 'Suspect', 'Pathological'], cmap = 'YlGnBu')
```

```python
from yellowbrick.classifier import ROCAUC

visualizerRandomForest = ROCAUC(randomForest, classes=['Normal', 'Suspect', 'Pathological'], macro=False)
visualizerRandomForest.fit(X_train, y_train)
visualizerRandomForest.score(X_test, y_test)
visualizerRandomForest.show()
```

## Tuned Decision Tree

```python
grid_params_dt={
    "criterion":["gini", "entropy"],
    "splitter":["best", "random"],
    "max_depth":[6, 8, 10, 12, 14],
    "min_samples_split":[2,4,6],
    "min_samples_leaf":[1,2]
}
```

```python
gs_dt = GridSearchCV(DecisionTreeClassifier(), grid_params_dt, verbose=1, n_jobs=-1)
gs_dt.fit(X_train, y_train)
```

```python
decisionTree = DecisionTreeClassifier(**gs_dt.best_params_).fit(X_train, y_train)
y_pred_decisionTree = decisionTree.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_decisionTree))
```

```python
print("Accuracy: ", accuracy_score(y_test, y_pred_decisionTree))
print("Precision: ", precision_score(y_test, y_pred_decisionTree, average="weighted"))
print("Recall: ", recall_score(y_test, y_pred_decisionTree, average="weighted"))
print("F1 Score: ", f1_score(y_test, y_pred_decisionTree, average="weighted"))
```

```python
plot_confusion_matrix(decisionTree, X_test, y_test, display_labels = ['Normal', 'Suspect', 'Pathological'], cmap = 'YlGnBu')
```

```python
visualizerDecisionTree = ROCAUC(decisionTree, classes=['Normal', 'Suspect', 'Pathological'], macro=False)
visualizerDecisionTree.fit(X_train, y_train)
visualizerDecisionTree.score(X_test, y_test)
visualizerDecisionTree.show()
```

# Tuned SVM

```python
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
 
grid = GridSearchCV(SVC(), param_grid, verbose=1, n_jobs=-1)
 
# fitting the model for grid search
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(classification_report(y_test, grid_predictions))
```

```python
print("Accuracy: ", accuracy_score(y_test, grid_predictions))
print("Precision: ", precision_score(y_test, grid_predictions, average="weighted"))
print("Recall: ", recall_score(y_test, grid_predictions, average="weighted"))
print("F1 Score: ", f1_score(y_test, grid_predictions, average="weighted"))
```

```python
plot_confusion_matrix(grid, X_test, y_test, display_labels = ['Normal', 'Suspect', 'Pathological'], cmap = 'YlGnBu')
```

```python
visualizerSVC = ROCAUC(grid, classes=['Normal', 'Suspect', 'Pathological'], macro=False)
visualizerSVC.fit(X_train, y_train)
visualizerSVC.score(X_test, y_test)
visualizerSVC.show()
```
